Part A

```python
import numpy as np
```

```python
fn="input04.txt"
```

```python
lines=[]
with open(fn,"r") as fin:
    text=fin.read()
```

```python
thrown_numbers_text, boards_text=text.split("\n\n",1)
boards_text=boards_text.split("\n\n")
boards_text=[board_text.split("\n") for board_text in boards_text]
```

```python
thrown_numbers=np.fromstring(thrown_numbers_text,sep=",",dtype=np.int)
```

```python
thrown_numbers
```

```python
def one_board_text_to_array(board_text):
    array_rows=[]
    for line_text in board_text:
        array_rows.append(np.fromstring(line_text,sep=" ",dtype=np.int))
    return np.array(array_rows)
```

```python
boards_text[0]
```

```python
boards=[one_board_text_to_array(board_text) for board_text in boards_text]
```

```python
boards
```

```python
trackers_status=[np.zeros((5,5)) for _ in range(len(boards))]
```

```python
# start playing
def update_game(boards, trackers_status,nr):
    for board,tracker_status in zip(boards,trackers_status):
        hit=(board==nr)
        tracker_status[hit]=1
    
def check_for_winner(boards,trackers_status):
    for board,tracker_status in zip(boards,trackers_status):
        check_rows=np.sum(tracker_status,1)
        check_cols=np.sum(tracker_status,0)
        if max(check_cols)==board.shape[0] or max(check_rows)==board.shape[0]:
            return board, tracker_status
```

```python
for nr in thrown_numbers:
    update_game(boards,trackers_status,nr)
    output=check_for_winner(boards,trackers_status)
    if output is not None:
        break
```

```python
# the winning board
board,tracker_status=output
```

```python
np.sum(board[tracker_status==0])*nr
```

Part B

```python
# reset game
trackers_status=[np.zeros((5,5)) for _ in range(len(boards))]
```

```python
is_finished=np.zeros((len(boards)))
```

```python
def update_who_finished(boards,trackers_status,is_finished):
    for idx,(board,tracker_status) in enumerate(zip(boards,trackers_status)):
        check_rows=np.sum(tracker_status,1)
        check_cols=np.sum(tracker_status,0)
        if max(check_cols)==board.shape[0] or max(check_rows)==board.shape[0]:
            is_finished[idx]=1
            
```

```python
for nr in thrown_numbers:
    update_game(boards,trackers_status,nr)
    update_who_finished(boards,trackers_status,is_finished)
    if np.sum(is_finished==0)<=1:
        break
```

```python
# this is the losing board
idx=int(np.where(is_finished==0)[0])

board=boards[idx]
tracker_status=trackers_status[idx]
```

```python
# now let's keep playing until that board is finished as well
losing_board=[board]
losing_tracker_status=[tracker_status]
```

```python
# a bit hacky: we start from the beginning again but doesn't matter
for nr in thrown_numbers:
    update_game(losing_board,losing_tracker_status,nr)
    output=check_for_winner(losing_board,losing_tracker_status)
    if output is not None:
        break
```

```python
output
```

```python
board,tracker_status=output
```

```python
np.sum(board[tracker_status==0])*nr
```
