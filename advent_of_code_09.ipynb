Part A

```python
import numpy as np
from skimage import morphology
import matplotlib.pyplot as plt
```

```python
with open ("input09.txt","r") as f:
    lines=f.readlines()
```

```python
for idx,line in enumerate(lines):
    line=line.strip("\n")
    line=[int(char) for char in line]
    lines[idx]=line
```

```python
img=np.array(lines)
plt.imshow(img)
```

```python
local_mins=(morphology.local_maxima(-img,connectivity=1))
```

```python
sum(1+img[local_mins])
```
