Part A

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fn="input05.txt"
```

```python
positions=[] # x1,y1,x2,y2
with open(fn,"r") as fin:
    for line in fin:
        line=line.strip("\n")
        line=line.replace(" ->","")
        line=line.replace(","," ")
        positions.append(np.fromstring(line,sep=" ",dtype=np.int))
```

```python
positions=np.array(positions)
```

```python
maxrows=max(np.max(positions[:,1]),np.max(positions[:,3]))+1
maxcols=max(np.max(positions[:,0]),np.max(positions[:,2]))+1
```

```python
danger_area=np.zeros((maxrows,maxcols),dtype=np.int)
danger_area.shape
```

```python
for position in positions:
    x1,y1,x2,y2=position
    # in one row (cst y)
    if y1==y2:
        xcoords=np.arange(min(x1,x2),max(x1,x2)+1)
        danger_area[y1,xcoords]+=1
    # in one col (cst x)
    elif x1==x2:
        ycoords=np.arange(min(y1,y2),max(y1,y2)+1)
        danger_area[ycoords,x1]+=1
        
```

```python
np.count_nonzero(danger_area>=2)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(danger_area)
```

Part B

```python
# reset the game
danger_area=np.zeros((maxrows,maxcols),dtype=np.int)
danger_area.shape
```

```python
for position in positions:
    x1,y1,x2,y2=position
    # in one row (cst y)
    if y1==y2:
        xcoords=np.arange(min(x1,x2),max(x1,x2)+1)
        danger_area[y1,xcoords]+=1
    # in one col (cst x)
    elif x1==x2:
        ycoords=np.arange(min(y1,y2),max(y1,y2)+1)
        danger_area[ycoords,x1]+=1
    # it's 45deg arranged
    else:
        xcoords=np.arange(min(x1,x2),max(x1,x2)+1)
        ycoords=np.arange(min(y1,y2),max(y1,y2)+1)
        # get direction correct
        if x2<x1:
            xcoords=np.flip(xcoords)
        if y2<y1:
            ycoords=np.flip(ycoords)
        
        danger_area[ycoords,xcoords]+=1
```

```python
np.count_nonzero(danger_area>=2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,10))
plt.imshow(danger_area)
```
